## Background and Prior Work


The World Cube Association (WCA) is the governing organization for Rubik's Cube competitions worldwide. The WCA recognizes 17 official events—ranging from 2x2, 3x3, ... 7x7, and other side events, such as Pyraminx, Megaminx, 3x3 blindfolded, 4x4 blindfolded, 5x5 blindfolded, etc—that can be held at the competition. While competitors are able to compete in any or all of these events, many typically choose to specialize in some subset of similar events - for instance, one competitor might focus on training blindfolded events (3x3 blindfolded, 4x4 blindfolded, and 5x5 blindfolded), whereas another competitor might specialize in big cubes (5x5, 6x6, and 7x7).  

In this project, we aim to analyze the correlation between results (average results) in each combination of the C(16, 2) events. Out of the 17, we remove "3x3 multiblind", an event which is scored based on the amount of 3x3 cubes a competitor can solve in an hour; this event uses "high score wins", whereas other events are scored on "low score wins". While analysis has been done on competition results in the past, most of it has focused on predicting future results in a certain event given one's previous times and improvement trend.<a id="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) By comparing an individual's results history in a specified event to other competitors at a given competition, one can reasonably predict where any competitor will rank in that competition as well as which competitors are most likely to break a world record at any given time. The only previous instance of analysis that seems to have been done on how results from different events correlate with eachother was centered on 3x3 times as a predictive benchmark for times in all other events.<a id="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Since 3x3 is the most popular and widely accessible event in which many of the fundamental skills used in solving other such twisty puzzles tend to be honed, cubers often like to draw correlations between certain time milestones in 3x3 to corresponding milestones in other events.

While this serves as a useful reference for how a competitior's times should look in the ideal scenario that they dedicate an equal amount of effort to every event, it provides no information on what a real life competitor's profile might look like if they did choose to practice a certain set of events more. Additionally, while it is the most commonly competed event, it is not gauranteed that any given cuber will have competed in 3x3 or that they will practice it enough to for the checkpoints to match up with their times in other events making only drawing comparisons to 3x3 results ineffective at showing how different competitors like to specialize. Utlimately, our project seeks to act less as a tool to track an individual's progress in an event and more as a reference tool for knowing how well sucess in one event correlates to sucess in any other event. This will be incredibly useful to competitors who are looking to branch out into trying new events as they will have an idea of which other events will reflect their particular solving interests the most based on what their favorite events. It will also benifit competition organizers as they will be better able to understand which events to host together to attract the most competitors as many cubers will not want to dedicate the time and money to travel to a competition that does not host multiple events that they are interested in.

# Hypothesis



We predict that the most closely related events would be between the 6x6 and 7x7 cubes.

Reasons we belive this is true is :
1) They have similar solution strategies and techniques:
  When solving a 6x6 or 7x7, both solutions use methods that solve the centers, then edges, then solve the rest like a 3x3. The similarities in solutions suggest that proficiency in one event would imply proficiency in the other, and as such, the times could therefore be closely correlated.
2) There's skill overlap amongst competitors of 6x6 and 7x7:
  Competitors who often do well in one big-cube event, like 6x6, are typically also interested in competing in other similar big-cube events, like 7x7. The skills developed while practicing one cube are directly transferable, such as finger dexterity and pattern recognition over multiple layers.
3) parity algorithms are similar:
  6x6 and 7x7 cubes have similar solving methods for parity algorithms. For all cubes larger than 3x3, there is a chance that that there's edge parity; to solve this, you can use a parity algorithm fore the last two edges. Competitors that know the algorithms for 6x6 may know and be good at the algorithms for 7x7, since they're the same..
4) Overlapping competition events:
  Competitions often host both big cube events, together, since competitors often have shared interests in both. Since their ranking is a function of if they compete and how many times they compete, cometitors that compete more in 6x6 will compete more in 7x7, and vice versa.
