# Research Question

-  Out of the 17 different Rubik's Cube competition events, which two events' results (speedcubing times), based on the World Cube Association (WCA) results database, are the most closely correlated?

## Background and Prior Work

The World Cube Association (WCA) is the governing organization for Rubik's Cube competitions worldwide. The WCA recognizes 17 official events—ranging from 2x2, 3x3, ... 7x7, and other side events, such as Pyraminx, Megaminx, 3x3 blindfolded, 4x4 blindfolded, 5x5 blindfolded, etc—that can be held at the competition. While competitors are able to compete in any or all of these events, many typically choose to specialize in some subset of similar events - for instance, one competitor might focus on training blindfolded events (3x3 blindfolded, 4x4 blindfolded, and 5x5 blindfolded), whereas another competitor might specialize in big cubes (5x5, 6x6, and 7x7).  

In this project, we aim to analyze the correlation between results (average results) in each combination of the C(16, 2) events. Out of the 17, we remove "3x3 multiblind", an event which is scored based on the amount of 3x3 cubes a competitor can solve in an hour; this event uses "high score wins", whereas other events are scored on "low score wins". While analysis has been done on competition results in the past, most of it has focused on predicting future results in a certain event given one's previous times and improvement trend.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) By comparing an individual's results history in a specified event to other competitors at a given competition, one can reasonably predict where any competitor will rank in that competition as well as which competitors are most likely to break a world record at any given time. The only previous instance of analysis that seems to have been done on how results from different events correlate with eachother was centered on 3x3 times as a predictive benchmark for times in all other events.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Since 3x3 is the most popular and widely accessible event in which many of the fundamental skills used in solving other such twisty puzzles tend to be honed, cubers often like to draw correlations between certain time milestones in 3x3 to corresponding milestones in other events. While this serves as a useful reference for how a competitior's times should look in the ideal scenario that they dedicate an equal amount of effort to every event, it provides no information on what a real life competitor's profile might look like if they did choose to practice a certain set of events more. Our project seeks to act less as a tool to track an individual's progress in an event and more as a reference for competition organizers to know what events will work well when hosted together or for competitiors looking to know where to branch out.

# Hypothesis


We predict that the most closely related events would be between the 6x6 and 7x7 cubes, because the bigger cubes have similar strategies to solve, and 6x6 and 7x7 are very similar in how they are solved. Additionally, if you are interested in solving 6x6, you are most likely interested in solving 7x7 and vice versa.
