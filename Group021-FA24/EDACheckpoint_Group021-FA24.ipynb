## Background and Prior Work


The World Cube Association (WCA) is the governing organization for Rubik's Cube competitions worldwide. The WCA recognizes 17 official events—ranging from 2x2, 3x3, ... 7x7, and other side events, such as Pyraminx, Megaminx, 3x3 blindfolded, 4x4 blindfolded, 5x5 blindfolded, etc—that can be held at the competition. While competitors are able to compete in any or all of these events, many typically choose to specialize in some subset of similar events - for instance, one competitor might focus on training blindfolded events (3x3 blindfolded, 4x4 blindfolded, and 5x5 blindfolded), whereas another competitor might specialize in big cubes (5x5, 6x6, and 7x7).  

In this project, we aim to analyze the correlation between results (average results) in each combination of the C(16, 2) events. Out of the 17, we remove "3x3 multiblind", an event which is scored based on the amount of 3x3 cubes a competitor can solve in an hour; this event uses "high score wins", whereas other events are scored on "low score wins". While analysis has been done on competition results in the past, most of it has focused on predicting future results in a certain event given one's previous times and improvement trend.<a id="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) By comparing an individual's results history in a specified event to other competitors at a given competition, one can reasonably predict where any competitor will rank in that competition as well as which competitors are most likely to break a world record at any given time. The only previous instance of analysis that seems to have been done on how results from different events correlate with eachother was centered on 3x3 times as a predictive benchmark for times in all other events.<a id="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Since 3x3 is the most popular and widely accessible event in which many of the fundamental skills used in solving other such twisty puzzles tend to be honed, cubers often like to draw correlations between certain time milestones in 3x3 to corresponding milestones in other events.

While this serves as a useful reference for how a competitior's times should look in the ideal scenario that they dedicate an equal amount of effort to every event, it provides no information on what a real life competitor's profile might look like if they did choose to practice a certain set of events more. Additionally, while it is the most commonly competed event, it is not gauranteed that any given cuber will have competed in 3x3 or that they will practice it enough to for the checkpoints to match up with their times in other events making only drawing comparisons to 3x3 results ineffective at showing how different competitors like to specialize. Utlimately, our project seeks to act less as a tool to track an individual's progress in an event and more as a reference tool for knowing how well sucess in one event correlates to sucess in any other event. This will be incredibly useful to competitors who are looking to branch out into trying new events as they will have an idea of which other events will reflect their particular solving interests the most based on what their favorite events. It will also benifit competition organizers as they will be better able to understand which events to host together to attract the most competitors as many cubers will not want to dedicate the time and money to travel to a competition that does not host multiple events that they are interested in.

# Hypothesis



We predict that the most closely related events would be between the 6x6 and 7x7 cubes.

Reasons we belive this is true is :
1) They have similar solution strategies and techniques:
  When solving a 6x6 or 7x7, both solutions use methods that solve the centers, then edges, then solve the rest like a 3x3. The similarities in solutions suggest that proficiency in one event would imply proficiency in the other, and as such, the times could therefore be closely correlated.
2) There's skill overlap amongst competitors of 6x6 and 7x7:
  Competitors who often do well in one big-cube event, like 6x6, are typically also interested in competing in other similar big-cube events, like 7x7. The skills developed while practicing one cube are directly transferable, such as finger dexterity and pattern recognition over multiple layers.
3) parity algorithms are similar:
  6x6 and 7x7 cubes have similar solving methods for parity algorithms. For all cubes larger than 3x3, there is a chance that that there's edge parity; to solve this, you can use a parity algorithm fore the last two edges. Competitors that know the algorithms for 6x6 may know and be good at the algorithms for 7x7, since they're the same..
4) Overlapping competition events:
  Competitions often host both big cube events, together, since competitors often have shared interests in both. Since their ranking is a function of if they compete and how many times they compete, cometitors that compete more in 6x6 will compete more in 7x7, and vice versa.

## Data Wrangling Explanation

Our dataset includes all official recorded solves from the first competition under the WCA on Jun 5, 1982 until November 12 2024 when the dataset was retrieved from the WCA database. The original dataset includes a lot of additional and unecessary information about each solve such as the name of the competitor, the name of the competition, and the round of the competition each solve was completed in. These have all been dropped as we do not need personal information about the competitor or the competition to make a judgement on the speed of the solve. The only such identifier we have kept is each competitor's WCA ID, a unique ID linked to each competitor, and is automatically generated when their results are first uploaded to the WCA database. This is the information we will be using to link multiple instances of competition solves to a competitor.

Additionally, we have chosen to drop a number of events that we have deemed irrelevant to the dataset whether it is, because they are no longer competed in or because they are too different to other events. 3x3 with feet, 3x3 old style multiblind, Rubik's Magic, and Rubik's Master Magic were a part of the dataset as they used to be officially recognized competition events but we have removed there since they are no longer officially recognized as a part of current WCA competitions and are thus no longer relevant to the goal of creating a useful reference for current competitors and organizers. Since these events have not been hosted official in a few years, it may not be accurate to consider 3x3 with feet, 3x3 old style multiblind, Rubik's Magic, and Rubik's Master Magic, as newer competitors may not have had a chance to compete in these events.

Furthermore, 3x3 Fewest Moves Challenge (FMC), and 3x3 Multiple Blindfolded (multi) have been removed because they are judged entirely differently and thus can not be compared with the methods we are using. FMC is judged on a lowest score metric similar to golf where the the result is a round integer representing the amount of moves a competitor took to solve a cube rather than the time they took. In this case, 3x3 Multi-blind  is judged by a high score metric where the amount of cubes unsuccessfully solved while blindfoled is subtracted from the number successfully solved while blindfolded instead of a time score. 3x3 Fewest Moves Challenge (FMC) is judgd based on the amount of moves it takes for a competitor to solve the cube from a given scramble. These alternative solving style events and their scoring metrics make it very hard to compare sucess in them to the other events that are judged on time as well as requiring very specific skills that are not very transferrable to other events. 

The remaining events we have left over are judged by either an average of 5 solves or a mean of 3 solves. The average of 5 is used for the majority of events and is calculated by dropping the fastest and slowest times then taking the mean of the three remaining times in an attempt to reduce the affect of lucky scrambles. The events that are just calculated by a mean of three tend to be longer events where the times are much more dependent on a solvers skill rather than how easy the scrambled state is to solve; as such taking the average of 5 (ao5) or mean of 3 (mo3) is dependent on the WCA regulations, which are written by the WCA Regulations Committee to standardize events for all cometitions. Both of these values are just stored under the 'average' column in the dataframe and are generally accepted to be comparable representaions of skill in the their events by the cubing community. As these are the standardized ways to measure success in each event. We have also dropped all the invidual time values and only kept the averages. We then dropped all the averages that were DNF (did not finish) and DNS (did not start) as they give us no information about the quality of the solves.

### Performance Patterns across all Events (EDA-1)

To better understand individual performance patterns in events held at WCA competitions, reorganized the dataset to be indexed by personID. Using personID as an identifier gets rid of inconsistensies that may arise by relying on name or any other non-unique column. To do so we created a new DataFrame by the name comp_avgs where each coloumn refers to the average score in each event. However because not everyone participates in every event there are many NaN values here. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

In order to help with plotting, first we used the following code to reorganize the dataframe to be indexed by personID, with each column having the average score in each event. Once again, note that there are a lot of NaN values here, as not everyone participates in every single event.

```python
comp_avgs = pd.DataFrame(df['personId']).set_index('personId')
for event in df['eventId'].unique():
    comp_avgs[event] = df[df['eventId'] == event].groupby('personId').mean('eventId')
```

```python
comp_avgs.head()
```

The column values here are in centiseconds. Each column represents a different event (e.g. 333 refers to a regular 3x3 cube solving event, 333oh refers to 3x3 one handed, 444 is a 4x4 Rubik's cube, etc.). And as explained above we can see how each row is identified by personID. 

Using this new dataframe that consolidates averages of each event, we generate several plots to observe potential correlations. We started with our hypothesis - the most closely related events would be between the 666 (6x6 cube) and the 777(7x7 cube) events. To look at this we plotted a scatter plot helping us gauge if there may be an initial correlation that is worth looking at.

```python
plt.scatter(comp_avgs['666'], comp_avgs['777'])
```

Another way we can look at correlating events is by looking at the following heat map plotted between every event. Events that that have a darker shade of red are the ones that are highly correlated, while darker blues are less correlated. It can be observed that 6x6 and 7x7 have a 0.92 coorelation, 5x5 and 6x6 have a 0.91 correlation, which are similar; comparatively, clock and 6x6 blindfolded have almost no correlation, at 0.039.

```python
correlations = comp_avgs.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Another way to sort these correlations is by generating each pair of events (17 choose 2), then sorting the event pair by correlation. From here, we can evaluate exactly whcih pairs of a higher correlation vs. lower correlation. In this case, it's clear that the two events with the lowest correlation, by far, are (clock and 4x4 blindfolded) and (clock and 5x5 blindfolded).

```python
event_pairs = []
correlations = []
for event1 in df['eventId'].unique():
    for event2 in df['eventId'].unique():
        if event1 != event2:
            if (event2, event1) not in event_pairs:
                event_pairs.append((event1, event2))
                correlations.append(comp_avgs[event1].corr(comp_avgs[event2]))
            

correlation_df = pd.DataFrame({'Event Pair': event_pairs, 'Correlation': correlations})
print(correlation_df.sort_values('Correlation', ascending=False))
```

Furthermore, we can graph the list of correlations between all pairs of events in a scatter plot to visualize the correlation between each pair of events. In particular, we can see if there are any pairs of events that have a particularly strong or particularly weak correlation.

```python
correlations = correlation_df['Correlation']

plt.figure(figsize=(10, 8))
correlations = correlations.sort_values()
plt.scatter(range(len(correlations)), correlations)
plt.xlabel('Event Pair')
plt.ylabel('Correlation')
plt.title('Correlation between event pairs')
plt.show()
```

### Competing Patterns across all Events (EDA-2)

Instead of comparing the correlations between the average times of competitors in two events, we can compare the Jaccard similarity between the competitors in those two events. The Jaccard similarity is a measure of how similar two sets are, and is calculated as the size of the intersection divided by the size of the union of the two sets. In this case, we can treat the competitors in each event as a set, and calculate the Jaccard similarity between the two sets of competitors.

In this case, we can treat the competitors in each event as a set, and calculate the Jaccard similarity between the two sets of competitors. We generate a table where it is 1 if they have a value >= 0, and 0 if they have a value of NaN. Then, we calculate the Jaccard similarity between the competitors in those two events.

```python
comp_avgs_binary = comp_avgs.copy()
comp_avgs_binary[comp_avgs_binary >= 0] = 1
comp_avgs_binary[comp_avgs_binary.isnull()] = 0
comp_avgs_binary = comp_avgs_binary.astype(int)
comp_avgs_binary.head()
```

Generate a list of all pairs of two events, and find the Jaccard similarity between the competitors in those two events.

```python
event_pairs_jaccard = []
jaccard_similarities = []
for event1 in df['eventId'].unique():
    for event2 in df['eventId'].unique():
        if event1 != event2:
            if (event2, event1) not in event_pairs_jaccard:
                event_pairs_jaccard.append((event1, event2))
                intersection = comp_avgs_binary[event1] & comp_avgs_binary[event2]
                union = comp_avgs_binary[event1] | comp_avgs_binary[event2]
                jaccard_similarities.append(intersection.sum() / union.sum())

jaccard_df = pd.DataFrame({'Event Pair': event_pairs_jaccard, 'Jaccard Similarity': jaccard_similarities})
jaccard_df.sort_values('Jaccard Similarity', ascending=False).head()
```

Similarly to the averages, we can also graph the list of correlations between all pairs of events in a scatter plot to visualize the Jaccard similarity between each pair of events. In particular, we can see if there are any pairs of events that have a particularly strong or particularly weak similarly.

```python
plt.figure(figsize=(10, 8))
jaccard_similarities = jaccard_df['Jaccard Similarity'].sort_values()
plt.scatter(range(len(jaccard_similarities)), jaccard_similarities)
plt.xlabel('Event Pair')
plt.ylabel('Jaccard Similarity')
plt.title('Jaccard Similarity between event pairs')
plt.show()
print(jaccard_df.sort_values('Jaccard Similarity', ascending=False))
```
